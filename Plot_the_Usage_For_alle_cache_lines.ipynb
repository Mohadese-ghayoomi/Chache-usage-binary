# **plot.py**

```
TSC_FREQUENCY = 2.4e9
DEF_REP = (0, 100)
DIR = "bin_search"
SIZE = (7, 5)
plt.rcParams.update({'font.size' : 12})
DIR = "bin_search"

server0 = sqlite3.connect("benchmark0.sqlite")
server1 = sqlite3.connect("benchmark1.sqlite")
server2 = sqlite3.connect("benchmark2.sqlite")

try:
    os.mkdir(DIR)
except:
    pass



def getAvgMeasurements(db, system, method, dataset, algorithm,
                       maxArraySize = (1<<32), repetitions = DEF_REP):
    c = db.cursor()
    c.execute("""SELECT size, avg(time), avg(iterations), avg(tsc), avg(inst_retired),
                 avg(clk_unhalted), avg(clk_unhalted_ref), avg(ctr0), avg(ctr1),
                 avg(ctr2), avg(ctr3) FROM measurements
                 WHERE system = (SELECT id FROM systems WHERE name = ?)
                 AND method = ?
                 AND dataset = (SELECT id FROM datasets WHERE name = ?)
                 AND algorithm = (SELECT id FROM algorithms WHERE name = ?)
                 AND repetition BETWEEN ? AND ?
                 GROUP BY size ORDER BY size""",
                 (system, method, dataset, algorithm,
                  repetitions[0], repetitions[1] - 1))

    sizes = list()
    measurements = {'time' : [], 'iterations' : [], 'tsc' : [], 'inst_retired' : [],
                    'clk_unhalted' : [], 'clk_unhalted_ref' : [],
                    'ctr0' : [], 'ctr1' : [], 'ctr2' : [], 'ctr3' : []}

    for size, time, iterations, tsc, inst_retired, clk_unhalted, clk_unhalted_ref, ctr0, ctr1, ctr2, ctr3 in c:
        if size > maxArraySize:
            break
        sizes.append(size)
        measurements['time'].append(time * 1.0e9)
        measurements['iterations'].append(iterations)
        measurements['tsc'].append(tsc)
        measurements['inst_retired'].append(inst_retired)
        measurements['clk_unhalted'].append(clk_unhalted)
        measurements['clk_unhalted_ref'].append(clk_unhalted_ref)
        measurements['ctr0'].append(ctr0)
        measurements['ctr1'].append(ctr1)
        measurements['ctr2'].append(ctr2)
        measurements['ctr3'].append(ctr3)

    return (sizes, measurements)


def getMeasurements(db, system, method, dataset, algorithm,
                    maxArraySize = (1<<32), repetitions = DEF_REP):
    c = db.cursor()
    c.execute("""SELECT size, repetition, time, iterations, tsc, inst_retired,
                 clk_unhalted, clk_unhalted_ref, ctr0, ctr1, ctr2, ctr3
                 FROM measurements
                 WHERE system = (SELECT id FROM systems WHERE name = ?)
                 AND method = ?
                 AND dataset = (SELECT id FROM datasets WHERE name = ?)
                 AND algorithm = (SELECT id FROM algorithms WHERE name = ?)
                 AND repetition BETWEEN ? AND ?
                 ORDER BY size, repetition""",
                 (system, method, dataset, algorithm,
                  repetitions[0], repetitions[1] - 1))

    sizes = list()
    measurements = {'time' : [], 'iterations' : [], 'tsc' : [], 'inst_retired' : [],
                    'clk_unhalted' : [], 'clk_unhalted_ref' : [],
                    'ctr0' : [], 'ctr1' : [], 'ctr2' : [], 'ctr3' : []}
    i = 0

    for size, repetition, time, iterations, tsc, inst_retired, clk_unhalted, clk_unhalted_ref, ctr0, ctr1, ctr2, ctr3 in c:
        if size > maxArraySize:
            break
        if len(sizes) == 0 or size > sizes[-1]:
            i += 1
            sizes.append(size)
            for x in measurements.values():
                x.append([])
        measurements['time'][-1].append(time * 1.0e9)
        measurements['iterations'][-1].append(iterations)
        measurements['tsc'][-1].append(tsc)
        measurements['inst_retired'][-1].append(inst_retired)
        measurements['clk_unhalted'][-1].append(clk_unhalted)
        measurements['clk_unhalted_ref'][-1].append(clk_unhalted_ref)
        measurements['ctr0'][-1].append(ctr0)
        measurements['ctr1'][-1].append(ctr1)
        measurements['ctr2'][-1].append(ctr2)
        measurements['ctr3'][-1].append(ctr3)

    return (sizes, measurements)


def calculate(sizes, measurements, value):
    if value in measurements:
        return measurements[value]
    elif value == 'keys_per_ns':
        return [x / y for x, y in zip(sizes, measurements['time'])]
    elif value == 'keys_per_clock':
        return [x / y for x, y in zip(sizes, measurements['clk_unhalted'])]
    elif value == 'ipc':
        return [x / y for x, y in zip(measurements['inst_retired'], measurements['clk_unhalted'])]
    elif value == 'branch_prediction_hit_ratio':
        return [1 - (x / y) for x, y in zip(measurements['ctr3'], measurements['ctr2'])]
    elif value == 'port0_utilization':
        return [x / y for x, y in zip(measurements['ctr0'], measurements['clk_unhalted'])]
    elif value == 'port1_utilization':
        return [x / y for x, y in zip(measurements['ctr1'], measurements['clk_unhalted'])]
    elif value == 'port2_and_3_utilization':
        return [x / y for x, y in zip(measurements['ctr2'], measurements['clk_unhalted'])]
    elif value == 'port5_utilization':
        return [x / y for x, y in zip(measurements['ctr3'], measurements['clk_unhalted'])]
    elif value == 'micro_ops_per_cycle':
        return [(a + b + c+ d) / cycles for a, b, c, d, cycles in zip(measurements['ctr0'], measurements['ctr1'], measurements['ctr2'], measurements['ctr3'], measurements['clk_unhalted'])]
    elif value == 'clk_diff':
        return [x - y for x, y in zip(measurements['clk_unhalted'], measurements['clk_unhalted_ref'])]
    elif value == 'clk_freq':
        return [TSC_FREQUENCY * (x / y) for x, y in zip(measurements['clk_unhalted'], measurements['clk_unhalted_ref'])]
    else:
        return 0


ylabels = {'time' : 'time in ns', 'keys_per_ns' : 'keys / ns', 'keys_per_clock' : 'keys / clock cycle',
           'iterations' : 'iterations', 'tsc' : 'time stamp counter',
           'inst_retired' : 'instructions retired', 'clk_unhalted' : 'unhalted clock cycles',
           'clk_unhalted_ref' : 'unhalted reference clock cycles',
           'ctr0' : 'LLC references', 'ctr1' : 'LLC misses',
           'ctr2' : 'branch instructions retired', 'ctr3' : 'branch misses retired',
           'ipc' : 'instructions per clock', 'branch_prediction_hit_ratio' : 'branch prediction hit ratio',
           'port0_utilization' : 'Port 0 utilization', 'port1_utilization' : 'Port 1 utilization',
           'port2_and_3_utilization' : 'Port 2 and 3 utilization', 'port5_utilization' : 'Port 5 utilization',
           'micro_ops_per_cycle' : 'Micro ops dispatched on port 0, 1, 2, 3 and 4 per cycle',
           'clk_diff' : 'clk_unhalted - clk_unhalted_ref', 'clk_freq' : 'frequency'}

default_styles = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'r--', 'g--', 'b--']


def plotLin(db, system, method, dataset, measurement, algorithms,
             labels = [], linestyles = default_styles, maxSize = 4096, repetitions = DEF_REP, markevery = 0.1, size = (12, 8)):
    fig, ax = subplots(figsize = size)
    for i, (algo, linestyle) in enumerate(zip(algorithms, linestyles)):
        sizes, values = getAvgMeasurements(db, system, method, dataset, algo, maxSize, repetitions)
        ax.plot(sizes, calculate(sizes, values, measurement), linestyle,
                label = labels[i] if i < len(labels) else algo, markevery = markevery)
    ax.set_xlabel('size in elements')
    ax.set_ylabel(ylabels[measurement])
    ax.set_xlim(0, ax.xaxis.get_data_interval()[1])
    ax.legend(loc='best')
    return fig


def plotLog(db, system, method, dataset, measurement, algorithms,
             labels = [], linestyles = default_styles, repetitions = DEF_REP, markevery = 0.1, size = (12, 8)):
    fig, ax = subplots(figsize = size)
    for i, (algo, linestyle) in enumerate(zip(algorithms, linestyles)):
        sizes, values = getAvgMeasurements(db, system, method, dataset, algo, repetitions=repetitions)
        ax.semilogx(sizes, calculate(sizes, values, measurement), linestyle, basex = 2,
                    label = labels[i] if i < len(labels) else algo, markevery = markevery)
    ax.set_xlabel('size in elements')
    ax.set_ylabel(ylabels[measurement])
    ax.set_xlim(0, ax.xaxis.get_data_interval()[1])
    ax.legend(loc='best')
    return fig


def joinMeasurements(sizes1, values1, sizes2, values2):
    resSizes = set(sizes1).intersection(set(sizes2))
    res1 = [v for s, v in zip(sizes1, values1) if s in resSizes]
    res2 = [v for s, v in zip(sizes2, values2) if s in resSizes]
    return sort(list(resSizes)), res1, res2


def calculateSpeedup(db, system, method, dataset, algo1, algo2, absolute=False, repetitions=DEF_REP):
    s1, m1 = getAvgMeasurements(db, system, method, dataset, algo1, repetitions=repetitions)
    s2, m2 = getAvgMeasurements(db, system, method, dataset, algo2, repetitions=repetitions)
    speed1 = calculate(s1, m1, 'time')
    speed2 = calculate(s2, m2, 'time')
    sizes, speed1, speed2 = joinMeasurements(s1, speed1, s2, speed2)
    speedup = 1 / (array(speed1) / array(speed2)) if not absolute else array(speed2) - array(speed1)
    return array(sizes), speedup


def plotSpeedupLin(db, system, method, dataset, algos,
                   linestyles = default_styles, markevery = 0.1, xrange = None, absolute=False,
                   repetitions = DEF_REP, size = (12, 8)):
    fig, ax = subplots(figsize=size)
    ax.set_xlabel('size in elements')
    ax.set_ylabel("speedup factor" if not absolute else "speedup in ns")
    for (algo1, algo2, label), linestyle in zip(algos, linestyles):
        ax.plot(*calculateSpeedup(db, system, method, dataset, algo1, algo2, absolute, repetitions), linestyle, markevery = markevery, label=label)
    if xrange == None:
        ax.set_xlim(0, ax.xaxis.get_data_interval()[1])
    else:
        ax.set_xlim(xrange)
    ax.legend(loc='best')
    return fig


def plotSpeedupLog(db, system, method, dataset, algos,
                   linestyles = default_styles, markevery = 0.1, xrange = None, absolute=False,
                   repetitions = DEF_REP, size = (12, 8)):
    fig, ax = subplots(figsize=size)
    ax.set_xlabel('size in elements')
    ax.set_ylabel("speedup factor" if not absolute else "speedup in ns")
    for (algo1, algo2, label), linestyle in zip(algos, linestyles):
        ax.semilogx(*calculateSpeedup(db, system, method, dataset, algo1, algo2, absolute, repetitions), linestyle, markevery = markevery, label=label, basex=2)
    if xrange == None:
        ax.set_xlim(0, ax.xaxis.get_data_interval()[1])
    else:
        ax.set_xlim(xrange)
    ax.legend(loc='best')
    return fig


def splitLegend(ax, firstLegendEntryCount, loc1 = 'upper left', loc2 = 'lower right'):
    l1 = ax.legend(handles=ax.get_lines()[:firstLegendEntryCount], loc=loc1)
    l2 = ax.legend(handles=ax.get_lines()[firstLegendEntryCount:], loc=loc2)
    ax.add_artist(l1)


def plotCacheSizes(ax):
    ax.axvline(8192, color='k', linestyle='--')
    ax.axvline(65536, color='k', linestyle='--')
    ax.axvline(5242800, color='k', linestyle='--')


def getAvgWarmUpMeasurements(db, system, method, dataset, algorithm):
    c = db.cursor()
    c.execute("""SELECT repetition, avg(time), avg(iterations), avg(tsc), avg(inst_retired),
                 avg(clk_unhalted), avg(clk_unhalted_ref), avg(ctr0), avg(ctr1),
                 avg(ctr2), avg(ctr3) FROM measurements
                 WHERE system = (SELECT id FROM systems WHERE name = ?)
                 AND method = ?
                 AND dataset = (SELECT id FROM datasets WHERE name = ?)
                 AND algorithm = (SELECT id FROM algorithms WHERE name = ?)
                 GROUP BY repetition ORDER BY repetition""", (system, method, dataset, algorithm))

    measurements = {'time' : [], 'iterations' : [], 'tsc' : [], 'inst_retired' : [],
                    'clk_unhalted' : [], 'clk_unhalted_ref' : [],
                    'ctr0' : [], 'ctr1' : [], 'ctr2' : [], 'ctr3' : []}

    for n, time, iterations, tsc, inst_retired, clk_unhalted, clk_unhalted_ref, ctr0, ctr1, ctr2, ctr3 in c:
        measurements['time'].append(time * 1.0e9)
        measurements['iterations'].append(iterations)
        measurements['tsc'].append(tsc)
        measurements['inst_retired'].append(inst_retired)
        measurements['clk_unhalted'].append(clk_unhalted)
        measurements['clk_unhalted_ref'].append(clk_unhalted_ref)
        measurements['ctr0'].append(ctr0)
        measurements['ctr1'].append(ctr1)
        measurements['ctr2'].append(ctr2)
        measurements['ctr3'].append(ctr3)

    return measurements


def getWarmUpMeasurements(db, system, method, dataset, algorithm):
    c = db.cursor()
    c.execute("""SELECT size, repetition, time, iterations, tsc, inst_retired,
                 clk_unhalted, clk_unhalted_ref, ctr0, ctr1, ctr2, ctr3
                 FROM measurements
                 WHERE system = (SELECT id FROM systems WHERE name = ?)
                 AND method = ?
                 AND dataset = (SELECT id FROM datasets WHERE name = ?)
                 AND algorithm = (SELECT id FROM algorithms WHERE name = ?)
                 ORDER BY size, repetition""", (system, method, dataset, algorithm))

    measurements = {'time' : [], 'iterations' : [], 'tsc' : [], 'inst_retired' : [],
                    'clk_unhalted' : [], 'clk_unhalted_ref' : [],
                    'ctr0' : [], 'ctr1' : [], 'ctr2' : [], 'ctr3' : []}

    i = -1
    for repetition, n, time, iterations, tsc, inst_retired, clk_unhalted, clk_unhalted_ref, ctr0, ctr1, ctr2, ctr3 in c:
        if repetition > i:
            i += 1
            for x in measurements.values():
                x.append([])
        measurements['time'][-1].append(time * 1.0e9)
        measurements['iterations'][-1].append(iterations)
        measurements['tsc'][-1].append(tsc)
        measurements['inst_retired'][-1].append(inst_retired)
        measurements['clk_unhalted'][-1].append(clk_unhalted)
        measurements['clk_unhalted_ref'][-1].append(clk_unhalted_ref)
        measurements['ctr0'][-1].append(ctr0)
        measurements['ctr1'][-1].append(ctr1)
        measurements['ctr2'][-1].append(ctr2)
        measurements['ctr3'][-1].append(ctr3)

    return measurements


def plotWarmUp(db, system, method, dataset, measurement, steps, stepSize,
               algorithms, labels = [], linestyles = default_styles, markevery = 0.1, size = (12, 8)):
    fig, ax = subplots(figsize = size)
    searches = [stepSize * x for x in range(0, steps)]
    for i, (algo, linestyle) in enumerate(zip(algorithms, linestyles)):
        values = getAvgWarmUpMeasurements(db, system, method, dataset, algo)
        ax.plot(searches, values[measurement], linestyle, markevery=markevery,
                label = labels[i] if i < len(labels) else algo)
    ax.set_xlabel('searches')
    ax.set_ylabel(ylabels[measurement])
    ax.set_xlim(0, ax.xaxis.get_data_interval()[1])
    ax.legend(loc='best')
    return fig

def readCacheUsageData(fileName):
    data = []
    with open(fileName) as file:
        for line in file:
            data.append(array(line.split(), dtype=int))
    return data

def plotCacheUsage(data, cacheLevel, size = (10, 6)):
    fig, ax = subplots(figsize = size)
    ax.plot(data[cacheLevel - 1])
    ax.set_xlabel('cache set')
    ax.set_ylabel('number of unique memory addresses')
    ax.set_title('Level {} Cache'.format(cacheLevel))
    ax.set_xlim(0, ax.xaxis.get_data_interval()[1])
    return fig

def plotCacheUsage2(ax, data, cacheLevel, label, linestyle):
    ax.plot(data[cacheLevel - 1], linestyle, label=label)
    ax.set_xlabel('cache set')
    ax.set_ylabel('number of unique memory addresses')
    ax.set_xlim(0, ax.xaxis.get_data_interval()[1])

```
